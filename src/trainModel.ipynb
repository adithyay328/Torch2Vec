```python
import sample
import model
import vectorize
import train
```

```python
import random
import os

def cleanReviewText(inText):
  replacements = [("<", ""), ("br", ""), ("/>", ""), ("""\'""", "'")]
  # Removing selected tags
  for rep in replacements:
    inText = inText.replace(rep[0], rep[1])
  
  return inText

# This function allows us to easily grab all text in a given
# review file. This function also handles text cleaning
def getText(fName=None, train=None, label=None):
  # If any of the params are none, that means the user
  # wants us to randomly select its value
  if train is None:
    train = random.choice([True, False])
  if label is None:
    label = random.choice([True, False])

  baseDir = "aclImdb"
  sourceDir = "train"
  if not train:
    sourceDir = "test"
  
  labelDir = "pos"
  if not label:
    labelDir = "neg"

  pathToFiles = f"./{baseDir}/{sourceDir}/{labelDir}"

  if fName is None:
    fileNames = os.listdir(pathToFiles)
    fName = random.choice(fileNames)
  
  with open(f"{pathToFiles}/{fName}", "r") as f:
    res = f.read()
    res = cleanReviewText(res)
    return res
```

```python
DOCCOUNT = 10
VOCABSIZE = 3000
vec = vectorize.Vectorizer(VOCABSIZE)
for i in range(1000):
    text = getText()
    vec.fit( [text] )

vec.limitVocab()
```

```python
vec.save("vectorizer.dump")
```

```python
myModel = model.SkipGramModel(VOCABSIZE, 300)
```

```python
EPOCHS = 15
for i in range(EPOCHS):
    corpus = []
    for i in range(DOCCOUNT):
        corpus.append(getText())
        
    sampler = sample.DocumentSampler(corpus, vec, 3)
    train.train(myModel, sampler)
```

```python
import torch
torch.save(myModel.state_dict(), "model_state_dict.dump")
```
